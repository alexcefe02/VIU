# Insertion Sort

```python
def insercion(A):
    print()
    print("Procedimiento:")
    comparaciones = 0
    intercambios = 0

    for i in range(len(A)):
        for j in range(i, 0, -1):
            comparaciones += 1
            if A[j - 1] > A[j]:
                A[j - 1], A[j] = A[j], A[j - 1]
                intercambios += 1
                print(A)

    print()
    print("Lista ordenada:", A)
    print("Comparaciones:", comparaciones)
    print("Intercambios:", intercambios)

# Programa Principal
A = [32, 8, 25, 15, 1, 6, 2, 3, 9, 190, 39943, 3, 7]
print("Lista original:", A)
insercion(A)
```

# Bubble Sort

```python
def bubbleSort(alist):
    print()
    print("Procedimiento:")
    comparaciones = 0
    intercambios = 0

    for passnum in range(len(alist) - 1, 0, -1):
        for i in range(passnum):
            comparaciones += 1
            if alist[i] > alist[i + 1]:
                # Intercambio
                alist[i], alist[i + 1] = alist[i + 1], alist[i]
                intercambios += 1
                print(alist)

    print()
    print("Lista ordenada:", alist)
    print("Comparaciones:", comparaciones)
    print("Intercambios:", intercambios)

# Programa Principal
alist = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("Lista original:", alist)
bubbleSort(alist)
```

# Selection Sort

```python
def selectionSort(alist):
    print()
    print("Procedimiento:")
    comparaciones = 0
    intercambios = 0

    for fillslot in range(len(alist)-1, 0, -1):
        positionOfMax = 0
        for location in range(1, fillslot+1):
            comparaciones += 1
            if alist[location] > alist[positionOfMax]:
                positionOfMax = location

        # Solo se intercambia si es necesario
        if positionOfMax != fillslot:
            alist[fillslot], alist[positionOfMax] = alist[positionOfMax], alist[fillslot]
            intercambios += 1
            print(alist)

    print()
    print("Lista ordenada:", alist)
    print("Comparaciones:", comparaciones)
    print("Intercambios:", intercambios)

# Programa Principal
alist = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("Lista original:", alist)
selectionSort(alist)
```

# Merge Sort 

```python
 
def mergeSort(alist):
    print("Procedimiento: ")
    def _mergeSort(alist):
        nonlocal comparaciones, copias
        print("Splitting ", alist)
        if len(alist) > 1:
            mid = len(alist) // 2
            lefthalf = alist[:mid]
            righthalf = alist[mid:]

            _mergeSort(lefthalf)
            _mergeSort(righthalf)

            i = 0
            j = 0
            k = 0

            while i < len(lefthalf) and j < len(righthalf):
                comparaciones += 1
                if lefthalf[i] < righthalf[j]:
                    alist[k] = lefthalf[i]
                    i += 1
                else:
                    alist[k] = righthalf[j]
                    j += 1
                copias += 1
                k += 1

            while i < len(lefthalf):
                alist[k] = lefthalf[i]
                i += 1
                k += 1
                copias += 1

            while j < len(righthalf):
                alist[k] = righthalf[j]
                j += 1
                k += 1
                copias += 1

            print("Merging ", alist)

    comparaciones = 0
    copias = 0
    _mergeSort(alist)
    print()
    print("Lista ordenada:", alist)
    print("Comparaciones:", comparaciones)
    print("Copias:", copias)

# Programa principal
alist = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("Lista original:", alist)
print()
mergeSort(alist)
```

# Heap Sort

```python
def heapify(arr, n, i, contadores):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    # Comparar con el hijo izquierdo
    if left < n:
        contadores["comparaciones"] += 1
        if arr[left] > arr[largest]:
            largest = left

    # Comparar con el hijo derecho
    if right < n:
        contadores["comparaciones"] += 1
        if arr[right] > arr[largest]:
            largest = right

    # Si el mayor no es la raíz
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        contadores["intercambios"] += 1
        print("Intercambio:", arr)
        heapify(arr, n, largest, contadores)


def heapSort(arr):
    print("Procedimiento: ")
    contadores = {"comparaciones": 0, "intercambios": 0}
    n = len(arr)

    # Construir el max heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i, contadores)

    # Extraer elementos del heap uno a uno
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]  # Mover raíz al final
        contadores["intercambios"] += 1
        print("Heap reducido:", arr)
        heapify(arr, i, 0, contadores)

    print("")
    print("Lista ordenada:", arr)
    print("Comparaciones:", contadores["comparaciones"])
    print("Intercambios:", contadores["intercambios"])

# Programa principal
arr = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("Lista original:", arr)
print()
heapSort(arr)
```

# Quick Sort

```python
def quickSort(arr):
    print("Procedimiento: ")
    
    contadores = {"comparaciones": 0, "intercambios": 0}

    def _quickSort(arr, low, high):
        if low < high:
            pi = partition(arr, low, high)
            _quickSort(arr, low, pi - 1)
            _quickSort(arr, pi + 1, high)

    def partition(arr, low, high):
        pivot = arr[high]
        i = low - 1  # Índice del menor elemento

        for j in range(low, high):
            contadores["comparaciones"] += 1
            if arr[j] <= pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
                contadores["intercambios"] += 1
                print("Intercambio:", arr)

        # Intercambiar el pivote con el siguiente elemento mayor
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        contadores["intercambios"] += 1
        print("Pivote colocado:", arr)
        return i + 1

    _quickSort(arr, 0, len(arr) - 1)
    print()
    print("Lista ordenada:", arr)
    print("Comparaciones:", contadores["comparaciones"])
    print("Intercambios:", contadores["intercambios"])

# Programa principal
arr = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("Lista original:", arr)
print()
quickSort(arr)
```

# Radix Sort

```python
def countingSort(arr, exp, contadores):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    # Contar ocurrencias en la posición `exp`
    for i in range(n):
        index = (arr[i] // exp) % 10
        count[index] += 1
        contadores["comparaciones"] += 1  # Se compara para contar

    # Acumular conteo (posiciones finales)
    for i in range(1, 10):
        count[i] += count[i - 1]

    # Construir salida estable
    i = n - 1
    while i >= 0:
        index = (arr[i] // exp) % 10
        output[count[index] - 1] = arr[i]
        count[index] -= 1
        contadores["copias"] += 1
        i -= 1

    # Copiar a arr[]
    for i in range(n):
        arr[i] = output[i]
        contadores["copias"] += 1

    print(f"Después de ordenar por posición {exp}: {arr}")


def radixSort(arr):
    print("Procedimiento: ")
    contadores = {"comparaciones": 0, "copias": 0}

    # Encuentra el número máximo para saber el número de dígitos
    max1 = max(arr)

    # Ordenar según cada dígito
    exp = 1
    while max1 // exp > 0:
        countingSort(arr, exp, contadores)
        exp *= 10

    print()
    print("Lista ordenada:", arr)
    print("Comparaciones (estimadas):", contadores["comparaciones"])
    print("Copias:", contadores["copias"])

# Programa principal
arr = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("Lista original:", arr)
print()
radixSort(arr)
```
