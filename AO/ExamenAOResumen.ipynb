# CAPÍTULO 1. INTRODUCCIÓN A LOS ALGORITMOS

## 1.1. ¿Qué es un algoritmo?
- Un algoritmo es un conjunto de reglas ordenadas y finitas para realizar una actividad mediante la
sucesión de pasos claramente establecidos.
- Creado por Al-Juarismi.

## 1.2. Tipos de algoritmos
- En funcion de la solución que ofrecen, deterministas y no deterministas
- En funcion del graod de complejidad y optimización: exactos, aproximados o heruísticos

## 1.3. Desarrollo e implementación de algoritmos
Distintas fases:
- Modelar
- Diseñar
- Analizar
- Optimizar
    - Si podemos establecer una función que permita esta comparación entre soluciones,
tendremos la función objetivo, para la que debemos encontrar la solución óptima entre todas las posibles. 

## 1.4. Complejidad computacional
Rendimiento del algoritmo. En la eficiencia del rendimiento se suelen
tener en cuenta dos elementos:
- el tiempo
- el espacio

### 1.4.1. Análisis de la complejidad de un algoritmo
Para ello comprobaremos el tiempo de ejecución real y el teorico
Usamos una medida teórica, que consiste en contar el número de instrucciones ejecutadas p

### 1.4.2. Órdenes de complejidad
- Orden constante, O(1).
- Orden logarítmico, O(log n).
- Orden lineal, O(n).
- Orden casi lineal, n · log n.
- Orden cuadrático, O(n2).
- Orden polinomial (a > 2), O(na).
- Orden exponencial (a > 1), O(an).
- Orden factorial, O(n!).


### 1.4.3. Tipo de problemas según su complejidad


- clase NP, para los que no se conoce un algoritmo que los resuelva en tiempo polinómico, pero para los que sí podemos verificar una solución en tiempo polinómico. 
-  resuelva en un tiempo polinomial, diremos que el problema es de clase P
